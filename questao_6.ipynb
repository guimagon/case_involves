### 6) Considere a instrução Python a seguir:
x = [ print(i) for i in range(10) if i % 2 == 0 ]
Após a execução dessa instrução no Python , a variável “x” conterá qual
valor.


#### Resposta:
Vc terá uma lista com 5 valores ‘none’: [None, None, None, None, None]
Isso acontece por o print sempre retorna none após a execução e a variável x só é atribuída depois que o print rodou, logo recebe o ‘None’.
Caso o print não existisse dentro da lista, e tivéssemos apenas x = [i for i in range(10) if i % 2 == 0 ], o retorno seria uma lista com os números pares [0, 2, 4, 6, 8]


```python
x = [ print(i) for i in range(10) if i % 2 == 0 ]
print(x)
```

```python
x = [ i for i in range(10) if i % 2 == 0 ]
print(x)
```
