#### 7) Faça um script em Python que peça dois números e imprima a soma.

```python
try:
    num_a = float(input('Digite o primeiro número (use ponto [.] como separador decimal): '))
    num_b = float(input('Digite o segundo número  (use ponto [.] como separador decimal): '))
    print(f'A soma dos números é: {num_a + num_b}')
except ValueError:
    print("Erro: a entrada não é numérica.")
```
